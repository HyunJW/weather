```python
import pandas as pd
```

```python
df1=pd.read_csv('./data/OBS_ASOS_TIM_20230323151247.csv', encoding="CP949")
df2=pd.read_csv('./data/OBS_ASOS_TIM_20230323151410.csv', encoding="CP949")
df3=pd.read_csv('./data/OBS_ASOS_TIM_20230323151612.csv', encoding="CP949")
df4=pd.read_csv('./data/OBS_ASOS_TIM_20230323151804.csv', encoding="CP949")
df5=pd.read_csv('./data/OBS_ASOS_TIM_20230323151958.csv', encoding="CP949")
```

```python
df=pd.concat([df1, df2, df3, df4, df5], axis=0)
```

```python
df
```

```python
est=pd.read_csv('./data/기상대기정보.csv')
```

```python
est
```

```python
# 일시 컬럼을 시계열 자료형으로 변경
df['일시'] = pd.to_datetime(df['일시'], format='%Y%m%d %H:%M')
est['일시'] = pd.to_datetime(est['일시'], format='%Y%m%d %H:%M')
```

```python
merge = pd.merge(est,df, how='left',on=['지점명','일시','지점'])
```

```python
merge
```

```python
merge.dropna(inplace=True)
```

```python
merge.to_csv('./data/기온대기정보_추가.csv', index=False)
```
