```python
df1=df[['PM10','PM2.5']]
```

```python
df1=df1.dropna()
```

```python
x=df1['PM10']
y=df1['PM2.5']
stats.linregress(x, y)
# PM2.5=-0.8556+0.6094*PM10
```

```python
x=df1['PM2.5']
y=df1['PM10']
stats.linregress(x, y)
# PM10=8.449+1.247*PM2.5
```

```python
# 미세, 초미세 농도 결측값 회귀추정하여 대체
df['PM10']=df['PM10'].fillna(8.449+1.247*df['PM2.5'])
df['PM2.5']=df['PM2.5'].fillna(-0.8556+0.6094*df['PM10'])
```

```python
df.to_csv('예측용관측소_수정.csv', index=False)
```
