```python
import os
os.chdir('C:/Users/tjoeun/Desktop/송수린/workspace/weather_project/data/예측용')
file_names=os.listdir()
```

```python
file_names
```

```python
columns = ['지점명', '일시', 'PM10', 'PM2.5', 'O3', 'NO2', 'CO', 'SO2']
excel = pd.DataFrame(columns=columns)
for observ in file_names:
    try:
        path = 'C:/Users/tjoeun/Desktop/송수린/workspace/weather_project/data/예측용/'
        files = glob.glob(path + f'{observ}')
        df = pd.read_excel(files[0], header=5)
        df = df[df.columns[0::2]]
        df.columns = ['일시', 'PM10', 'PM2.5', 'O3', 'NO2', 'CO', 'SO2']
        df['일시'] = df['일시'].apply(lambda x: (x.replace(':', ' ') + ':00'))
        df['지점명'] = observ.split(' ')[0]
        excel = pd.concat([excel, df], axis=0)
        
        excel.sort_values('일시', inplace=True)
        excel.to_csv(path + '예측용관측소.csv', index=False)

    except Exception as ex:
        print('error' + str(ex))
```
