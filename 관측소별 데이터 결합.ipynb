```python
import pandas as pd
import glob
```

```python
df_addr = pd.read_csv('./data/관측소주소.csv', encoding='cp949')
```

```python
obsevatory = df_addr['관측소명'].apply(lambda x: x.split('(')[0])
```

```python
cnt=0
for observ in obsevatory:
    try:
        path = 'C:/Users/tjoeun/Desktop/관측소/'
        files = glob.glob(path + f"{cnt}.{observ} 관측소/" + "*.xls")
        cnt += 1
        excel = pd.DataFrame()
        for file_name in files:
            df = pd.read_excel(file_name, header=5)
            df = df[df.columns[0::2]]
            df.columns = ['일시', 'PM10', 'PM2.5', 'Ozon', 'NO2', 'CO', 'SO2']
            excel = excel.append(df, ignore_index=True)
        excel['일시'] = excel['일시'].apply(lambda x: (x.replace(':', ' ') + ':00'))
        excel.sort_values('일시', inplace=True)
        excel['지점명'] = observ
        columns = ['지점명', '일시', 'PM10', 'PM2.5', 'Ozon', 'NO2', 'CO', 'SO2']
        excel = excel[columns]
        excel.to_csv(path + '관측소별 결합 데이터/' + f'{observ}관측소2018~2022.csv', index=False)

    except Exception as ex:
        print('error' + str(ex))
```
